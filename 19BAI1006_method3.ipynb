```python
count = 0
def domainAge(domain_name):   
    global count
    
    try:
        res=whois.whois(domain_name)
        creation_date = res.creation_date[0]
        expiration_date=res.expiration_date[0]
        if (expiration_date.year - creation_date.year ) < 2 :
            print(str(count) + ") may be phising site...")
            count +=1
            return 1
        else:
            print( str(count)+ ") may be legitmate site...")
            count +=1
           
            return 0
    except:
    
        
        print(str(count) + ") as the site is not there in database .....")
        count +=1
        return 1
   
```

```python
df['dom_check']=df['domain_name'].apply(lambda i : domainAge(i))
```
